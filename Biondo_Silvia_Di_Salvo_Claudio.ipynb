## Biogas plants location

An association of $n$ farmers wants to open $p$ plants to produce energy from biogas. 
Each plant will be opened at a farm of a member of the association and will be powered with corn chopping purchased from the farm itself or from other neighboring farms.

Each farm $i$ can provide at most $c_i$ tons of corn chopping, with a percentage of dry matter $a_i$. As you may know, dry matter is the key component of corn chopping used for biogas production. In order to maintain the quality of produced energy, each plant must burn a mixture of corn chopping with a percentage of dry matter between $k_{min}$ and $k_{max}$. 

At most one plant can be located in each farm, and every farm can sell its corn chopping to one and only one plant.

Each farm $i$ is located at coordinates $x_i$ and $y_i$, representing respectively its latitude and longitude, and the cost of moving corn chopping from a farm $i$ to a farm $j$ is proportional to the euclidean distance between the two farms (it does not depend on the actual quantity moved, since the trucks used for this transportations are sufficiently big). 

Under such conditions, every plant produces $Q$ kWh of energy per ton of corn chopping burned. The energy produced by each plant will be fed into the national electricity system, at a unitary price of $b$ (€/kWh). Moreover, due to state regulations, each plant must not produce more than $M$ kWh of energy.

You must locate $p$ plants among the available farms and assign the farms that will supply each plant, with the goal of maximizing the total revenues of the association.

### Sets
*   $I$ = set of farms

### Parameters
*   $n$ = number of farms   
*   $p$ = number of plants to locate
*   $b$ = revenue per unit of energy (€/kWh)
*   $M$ = max energy production (kWh)
*   $Q$ = energy produced by a ton of corn chopping (kWh/t)
*   $k_{min} (k_{max})$ = min (max) percentage of dry matter for fermentation
*   $a_i$ = percentage of dry matter in chopping from farm $i \in I$
*   $c_i$ = tons of corn chopping available for each $i \in I$ (t)
*   $x_i, y_i$ = coordinates of farm $i \in I$

```python
from traitlets.traitlets import Instance
import json
import mip
import numpy as np
import math
from pandas import *
import time
import csv
from google.colab.data_table import DataTable
DataTable.max_columns = 500


def print_result(m,x,y,c_crd,n,p):
    import networkx as nx
    import matplotlib.pyplot as plt
    
    print(f"Print_Result objective_value: {m.objective_value:.2f}")
    f_crd=[[0 for i in range(n)]for j in range(n)]
    for j in range(n):
        if((c_crd[j][0]*y[j].x) > 0):
          f_crd[j][0] = c_crd[j][0]
          f_crd[j][1] = c_crd[j][1]

    print("Plant:")
    for j in range(n):
        if y[j].x > 0+1e-8:
            print(f"{j:10d} --> {y[j].x:7.3f}; Farms: ", end='')
            for i in range(n):
                if x[i,j].x > 0+1e-8:
                    print(f"{i} ({x[i,j].x:.2f}); ", end='')
            print('')
    # Visualize solution on graph

    pos_f = {f'P{j}': (f_crd[j][0], f_crd[j][1]) for j in range(n)}
    pos_c = {f'F{i}': (c_crd[i][0], c_crd[i][1]) for i in range(n)}
    
    nodepos = {**pos_f, **pos_c}
    
    g = nx.Graph()
    g.add_nodes_from([f'P{j}' for j in range(n)])
    g.add_nodes_from([f'F{i}' for i in range(n)])
    edges = [(f'P{j}', f'F{i}') for j in range(n) for i in range(n) if x[i,j].x > 1e-8] 
    g.add_edges_from(edges)
    color = ['yellow' if y[j].x !=1 else 'cyan' for j in range(n)] + ['pink']*n
    plt.figure(1, figsize=(14,14))
    nx.draw_networkx(g, font_size=12, pos=nodepos, node_color=color)
    plt.show()
    
# Reads a .json instance and returns it in a dictionary
def load_instance(filename):
  with open(filename, 'r') as f:
    data = json.load(f)
  return data

# Reads a .txt result and returns it
def load_result(filename):
  with open(filename, 'r') as f:
      result = f.read()
  return float(result)

def solve(instance):
  
  m = mip.Model()
  n = instance['n']
  a = instance['a']
  c = instance['c']
  c_crd = instance['points']
  p = instance['p']

  x = [m.add_var(var_type=mip.CONTINUOUS, lb = 0)for i in range(instance['n'])]
  kw = m.add_var(var_type=mip.CONTINUOUS) 
  farm =  [m.add_var(var_type=mip.CONTINUOUS,ub = 1, lb=0)for i in range(instance['n'])] #FARM[I] = 1 SE PRESENTE UN PLANT NELLA J-ESIMA FARM

  presente =  { (i, j) : m.add_var(var_type=mip.CONTINUOUS, ub = 1, lb=0) for i in range(n) for j in range(n) } #RAPPRESENTA QUALE FARM DA A QUALE PLANT
  #SE PRESENTE[5][8] = 1 SIGNIFICA CHE IL PLANT 8 RICEVERà DALLA FARM 5
  #LA SOMMA DI UNA COLONNA SARà PARI AL NUMERO DI FARM CHE RIFORNIRANNO UN PLANT J


  #CONSTRAINT SULLA DRY MATTER
  m.add_constr((mip.xsum(x[i]*a[i] for i in range(n)) ) >= mip.xsum(x[i]  for i in range(n))* instance['kmin'])
  m.add_constr((mip.xsum(x[i]*a[i] for i in range(n)) ) <= mip.xsum(x[i]  for i in range(n))* instance['kmax'])
  

  for i in range(n):  
      m.add_constr(x[i]<= c[i]) #IL CORN CHE COMPRO DEV'ESSERE INFERIORE DA QUELLO DISPONIBILE
 

  m.add_constr(mip.xsum(farm[i] for i in range(n))==instance['p'])#GARANTIAMO IL NUMERO DI PLANT

  #CONSTRAINT SELL PLANT
  for i in range(n):  
      m.add_constr(mip.xsum(presente[i,j] for j in range(n)) == 1) #UNA FARM PUò FORNIRE UNA FARM
  for j in range(n):
      m.add_constr(mip.xsum(presente[i,j]  for i in range(n)) <= n * farm[j]  ) # UN PLANT PUò ESSER RIFORNITO DA UN MASSIMO DI n FAMR


 #CONSTRAINT SULLA PRODUZIONE DELL'ENERGIA
  kw = mip.xsum((x[i]  * instance['Q'] ) for i in range(n))
  m.add_constr(kw<=instance['M']*instance['p'])


  #CALCOLO DISTANZA EUCLIDEA
  move_cost = {(i,j): 1 * math.sqrt((c_crd[i][0] - c_crd[j][0])**2 +
                                    (c_crd[i][1] - c_crd[j][1])**2)  for i in range(n) for j in range(n)}


  m.objective = ( mip.maximize(kw* instance['b']) - 
                  mip.minimize(mip.xsum((presente[i,j] * move_cost[i,j]) for i in range(n) for j in range(n) )) )
  
  m.optimize()
  


#CUTTING PLANE
  eps = 1e-5
  #USIAMO IL CUTTING PLANE AL FINE DI RISOLVERE LA LP RELAXATION CHE SI CREA SULLE VARIABILI PRESENTE E FARM
  while True:
    pair = None  # Initialize pair to None, leave the loops
    m.optimize()
    for i in range(n):
        for j in range(n):
            # Check if inequality is violated (Add an epsilon to avoid numerical problems)
            if presente[i,j].x > farm[j].x + eps:
                pair = (i,j)
                break  # Leave the inner loop
            if farm[j].x == 0.5:
                m.add_constr(presente[i,j]==farm[j])
                pair = (-1,-1)
                break;
        if pair != None:
            break  # Leave the outer loop if an inequality is found

    if pair == None:
        break  # No violated inequality was found, leave the loop

    i,j = pair

    # Add violated constraint
    if(i!=-1) and (j!=-1):
      m.add_constr(presente[i,j] <= farm[j])
    
  return m.objective_value

  
f = open('result.csv', 'w')
header = ['ID', 'OBJ', 'TIME']
writer = csv.writer(f)
writer.writerow(header)
f.close()
for i in range(1,11):
  start = time.time()
  inst = load_instance("instance_"+str(i)+".json")
  #res = load_result("instance_3.txt")

  obj = solve(inst)
  #gap = 100 * (obj - res) / res

  print("result: {}".format(obj))
  #print("expected: {}".format(res))
  #print("gap: {}".format(gap))
  end = time.time()
  f = open('result.csv', 'a')
  data = [i,obj,end-start]
  writer = csv.writer(f)
  writer.writerow(data)
  f.close()
```
